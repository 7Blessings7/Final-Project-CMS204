```python
from google.colab import files
import csv

class Graph:
    def __init__(self, edges):                                          #("Rosario", "Tejero", 2.1)
        self.edges = edges
        self.graph_dict = {}                                            #creates the dictionary for the graph, each key (start location) is assigned a tuple that the end location and distance
        for start, end, dist in edges:
            if start in self.graph_dict:
                self.graph_dict[start].append((end,dist))               #creates key from the start node, creates a tuple for the correspoding value {"Rosario" : ("Tejero", 2.1)}
            else:
                self.graph_dict[start] = [(end,dist)]

    def get_routes(self, start, end, path=[]):                          #creates a list of routes from given start location to end location
        path = path + [start]                                           #updates path

        if start == end:
            return [path]

        if start not in self.graph_dict:
            return []

        paths = []                                                      #initializes paths which is the output list
        for node in self.graph_dict[start]:                             #checking each node in the dictionary
            if node not in path:                                        #setup recursion
                new_paths = self.get_routes(node[0], end, path)         #node modified to node[0] to get the location part of the tuple, since the dictionary was modified to contain tuples
                for p in new_paths:
                    paths.append(p)                                     #appends output list with the obtained paths
        return paths

    def get_distance(self, route):                                      #uses the list of paths to get the corresponding total distance of the path, modifies the list to contain tuples of the visited nodes and corresponding
        dist = 0                                                        #total distance travelled
        new_list = []                                                   #initializes the distance and the new list of tuples
        for i in route:
            if i == route[-1]:                                          #conditional to serve as stopping point, prevents the program from iterating using the end point which is unnecessary
                return new_list
            for node in self.graph_dict[i]:                             #checks the dictionary for the destinations available from the given location
                location = node[0]                                      #gets location part of tuple
                if route.index(i) + 1 < len(route):                     #prevents out of index error
                    if location == route[route.index(i) + 1]:           #checks if the obtained destination is the destination specified in the path
                        dist = dist + node[1]                           #updates the distance travelled if true
                        new_list.append((location, dist))               #appends the list with a tuple if true
        return new_list

    def print_routes(self, start, end):                                 #provides a formatted output for the routes with distances travelled
      gp = self.get_routes(start,end)                                   #calls the get routes method
      master_list = []                                                  #initializes the new list for output ---- from get routes paths list to modified list with destination, distance tuples using get_distance method 
      sr = []                                                           #initializes the list for shortest path
      for i in gp:                                                      
        if isinstance(i,list):                                          #checks if the list output from get routes method contains lists of routes 
          master_list.append(self.get_distance(i))                      #feeds the lists one by one to get distance method if true 
        else:
          master_list.append(self.get_distance(gp))                     #if the list output from get routes is not a list, then feeds that output to get distance method 
      min = float('inf')
      if len(master_list) == 0:                                          #checks if new list for output is empty and provides notice that no routes were found if true
        print("No routes found for the given locations")                
      else:                                                             #if list is not empty
        if len(master_list) == 1:                                       #checks if the list contains only one path to prevent index error
          if master_list[0] == []:                                      #checks if new list output contains an empty set and provides notice that no routes were found if true
            print("No routes found for the given locations")
          else:  
            min = master_list[-1][-1][-1]                               #gets the min distance for non-empty 1 item new list output
            sr = master_list[0]                                         #sets the only item in the list as the shortest route 'sr'
        if len(master_list) > 1:  
          for x in range (len(master_list)):                            #iterates over the distances in new list outputs with 2 items or more
            f = master_list[x][-1][-1]                                  #uses -1 index to get the distance travelled
            if min > f:                                                 #compares minimum
              min = f                                                   #updates minimum
              sr = master_list[x]                                       #records path with min distance travelled

        print("The routes from ",start, "to ",end, "are the following:")#formats the output of the printout
        for i in range(len(master_list)):
          print("Route #",i+1)
          print("Location",'\t',"Distance travelled")
          if len(start) < 7:
            print(start,'\t','\t',"---")
          else:
            print(start,'\t',"---")
          for i in master_list[i]:
            if len(i[0]) < 7:
              print(i[0],'\t','\t',i[1],"km")
            else:
              print(i[0],'\t',i[1],"km")
          print()
        print("The shortest route with a distance of", min,"is:")
        print("Location",'\t',"Distance travelled")
        if len(start) < 7:
          print(start,'\t','\t',"---")
        else:
          print(start,'\t',"---")
        for i in sr:
            if len(i[0]) < 7:
              print(i[0],'\t','\t',i[1],"km")
            else:
              print(i[0],'\t',i[1],"km")
      return
```

```python
def add_edge_set():                                               #uses an uploaded file with format of columns, 1st = start, 2nd = end, 3rd = distance
  edge_set = []                                                   #initializes list to be used to create the graph
  locations = []                                                  #initializes the list of available locations for user's convenience
  uploaded = files.upload()                                       #prompts file upload
  name = list(uploaded.keys())                                    #records name of file
  with open(name[0], 'r', encoding = 'utf-8-sig') as file:        #uses csv reader to populate the edge_set with lists in the prescribed format
    reader = csv.reader(file)
    for row in reader:
      edge_set.append((row[0], row[1], float(row[2])))            #portion that appends the list for the graph
      if row[0] not in locations:                                 #portion that the list of available locations
        locations.append(row[0])
      if row[1] not in locations:
        locations.append(row[1])

  print("-----------------------------------------------")
  print("These are the available locations:")                     #sets printout
  print(locations)                                                #prints available locations
  return edge_set                                                 #returns the set to be used to create the graph

if __name__ == '__main__':

  edge_set = add_edge_set()                                       #calls add_edge_set and stores the set to the variable
  test = Graph(edge_set)                                          #creates the graph
  print("These are the available routes")
  print(test.graph_dict)                                          #prints the dictionary
```
