```python
import networkx as nx

# Create the graph
G = nx.Graph()

# Add employees to the graph as nodes
for employee in employeeData['employees']:
    G.add_node(employee['name'],department=employee['department'], sub_dept=employee['sub_dept'], team=employee['team'])

# Add edges to the graph based on the strength of connection between employees
for i in range(len(employeeData['employees'])):
    for j in range(i+1, len(employeeData['employees'])):
        weight = 0
        employee1 = employeeData['employees'][i]
        employee2 = employeeData['employees'][j]
        if employee1['team'] == employee2['team']:
            weight = 3
        elif employee1['sub_dept'] == employee2['sub_dept']:
            weight = 2
        elif employee1['department'] == employee2['department']:
            weight = 1
        if weight > 0:
            G.add_edge(employee1['name'], employee2['name'], weight=weight)


# Create functions that generate lists containing employees based on strength of connection
def find_primary_connections(node):
    primaryConnections = []
    for n, w in G[node].items():
        if w['weight'] == 3:
            primaryConnections.append(n)
    return primaryConnections

def find_secondary_connections(node):
    secondaryConnections = []
    for n, w in G[node].items():
        if w['weight'] == 2:
            secondaryConnections.append(n)
    return secondaryConnections

def find_tertiary_connections(node):
    tertiaryConnections = []
    for n, w in G[node].items():
        if w['weight'] == 1:
           tertiaryConnections.append(n)
    return tertiaryConnections

# Print out 3 lists for each employee in the deparment (1 list each contraining strong, medium, and weak connections)
for employee in employeeData['employees']:
    print(f"{employee['name']} ({employee['sub_dept']} - {employee['team']}) - Primary: {find_primary_connections(employee['name'])}, Secondary: {find_secondary_connections(employee['name'])}, Tertiary: {find_tertiary_connections(employee['name'])}")
```
