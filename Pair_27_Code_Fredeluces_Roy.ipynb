```python
class Graph:
  def __init__(self, num_of_nodes):
    self.m_num_of_nodes = num_of_nodes
    self.list_of_activities = []
        
    # Different representations of a graph
    self.m_list_of_edges = []
	
  def add_edge(self, node1, node2, weight=1, price=1):        
    # Add the edge from node1 to node2
    self.m_list_of_edges.append([node1, node2, weight, price])

    self.list_of_activities.append(node1) if node1 not in self.list_of_activities else None

  def print_edge_list(self):
    num_of_edges = len(self.m_list_of_edges)
    for i in range(num_of_edges):
        print("edge ", i+1, ": ", self.m_list_of_edges[i])

class BinarySearchTreeNode:
  def __init__(self,data, mode):
    self.data = data
    self.left = None
    self.right = None
    self.mode = mode
    if mode == "price":
      self.idx = 3
    elif mode == "review":
      self.idx = 2
  
  def add_child(self, data):
    if data == self.data: # checker for if data is already existing in the set then, do not add
      return
    
    if data[self.idx] < self.data[self.idx]: # add this value to left subtree
      if self.left: # check if this is not a leaf node
        self.left.add_child(data)
      else:
        self.left = BinarySearchTreeNode(data, self.mode)
    else: #if data is greater then add to R subtree
      if self.right: # check if this is not a leaf node
        self.right.add_child(data)
      else:
        self.right = BinarySearchTreeNode(data,self.mode)
  
  def in_order_traversal(self, mode = "price"):
    elements = [] #empty list for elements to append to while travesring

    if mode == 'review':
      #visit right tree
      if self.right:
        elements += self.right.in_order_traversal(self.mode)

      #visit base node
      elements.append(self.data)

      # visit left node
      if self.left:
        elements += self.left.in_order_traversal(self.mode)
    elif mode == 'price':
      # visit left node
      if self.left:
        elements += self.left.in_order_traversal(self.mode)
      
      #visit base node
      elements.append(self.data)

      if self.right:
        elements += self.right.in_order_traversal(self.mode)

    return elements

class SuggestedActivities:
  def __init__(self, mode, list_of_activities, graph):
    self.mode = mode
    self.list_of_activities = list_of_activities
    self.tree = self.build_tree(graph.m_list_of_edges)
    
    self.activities = self.tree.in_order_traversal(mode)

    # print(self.activities)


  def build_tree(self, elements):
    root = BinarySearchTreeNode(elements[0], self.mode)

    for i in range(1, len(elements)):
      root.add_child(elements[i])

    return root

  def filter_acts(self, act,x):
    return True if act[0] == x else False

  def suggest_activity(self, act, al=[]):
    act_list = al

    if act not in self.list_of_activities:
      if act not in act_list:
        act_list.append(act)
      else:
        pass
      return act_list

    if act in self.list_of_activities:
      if act not in act_list:
        act_list.append(act)
      suggest = [x for x in self.activities if self.filter_acts(x, act)] #Filter activities based on current activity

      if suggest: 
        next_activity = suggest[0][1] # Get the next activity based on the weight of the relation

        act_list.append(next_activity) #append the the new activity to act_list
        act = next_activity #assign the next activity as the current activity
        self.suggest_activity(act, act_list)
      else:
        return act_list.append(act)
    
    return act_list

g = Graph(5)

g.add_edge("Movie", "Dinner", 25, 15)
g.add_edge("Movie", "Walk in Park", 28, 25)
g.add_edge("Movie", "Ice Cream", 3, 29)
g.add_edge("Walk in Park", "Fireworks", 7, 11)
g.add_edge("Walk in Park", "Ice Cream", 8, 4)

list_of_activities = g.list_of_activities

recommendations = SuggestedActivities("review", list_of_activities, g) #"review" or "price"

print(recommendations.suggest_activity("Movie"))
```
