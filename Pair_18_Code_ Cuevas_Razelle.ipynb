```python
#base code: https://github.com/codebasics/data-structures-algorithms-python/blob/master/data_structures/10_Graph/graph.py
import numpy
class Graph:
  def __init__(self, edges):
    self.edges = edges
    self.graph_dict = {}
    for start, end, dist, time, price in edges:
      if start in self.graph_dict:
        if end in self.graph_dict[start]:
          self.graph_dict[start][end].update([dist, time, price])
        else:
          self.graph_dict[start].update({end: [dist, time, price]})
      else:
        self.graph_dict[start] = {end: [dist, time, price]}
    
    for item in self.graph_dict:
      print(item)
      for point in self.graph_dict[item]:
        details = "[ Distance: " + str(self.graph_dict[item][point][0]) + " | "
        details += "Time: " + str(self.graph_dict[item][point][1]) + " | "
        details += "Price: " + str(self.graph_dict[item][point][2]) + " ]"
        print("   " + point + " -> " + details) 

  def get_paths(self, start, end, path=[], amount=[]):
    path = path + [start]
    amount = self.compute(path)

    if start == end:
      return [path], [amount]

    if start not in self.graph_dict:
      return [], []

    paths = []
    amounts = []
    for node in self.graph_dict[start]:
      if node not in path:
        new_paths = self.get_paths(node, end, path, amount)
        cnt = 0
        for p in new_paths[0]:
          paths.append(p)
          amounts.append(new_paths[1][cnt])
          if(cnt < len(new_paths[0]) - 1):
            cnt += 1

    return paths, amounts
  
  def compute(self, paths):
    cnt = 0
    dist = 0
    time = 0
    price = 0

    for path in paths:
      if(cnt >= 1):
        dist += self.graph_dict[paths[cnt-1]][path][0]
        time += self.graph_dict[paths[cnt-1]][path][1]
        price += self.graph_dict[paths[cnt-1]][path][2]
      cnt += 1
    return [dist ,time, price]
  
  def get_shortest_and_longest(self, paths):
    node_count = []
    for i in paths[0]:
      node_count.append(len(i))
    
    least = min(node_count)
    most = max(node_count)

    arr1 = numpy.array(node_count)
    shortest = numpy.where(arr1 == least)[0]
    arr2 = numpy.array(node_count)
    longest = numpy.where(arr2 == most)[0]

    print("\nThe shortest path(s):")
    for i in shortest:
      details = "[ Distance: " + str(paths[1][i][0]) + " | "
      details += "Time: " + str(paths[1][i][1]) + " | "
      details += "Price: " + str(paths[1][i][2]) + " ]"
      print(str(paths[0][i]) + " -> " + details)

    print("\nThe longest path(s):")
    for i in longest:
      details = "[ Distance: " + str(paths[1][i][0]) + " | "
      details += "Time: " + str(paths[1][i][1]) + " | "
      details += "Price: " + str(paths[1][i][2]) + " ]"
      print(str(paths[0][i]) + " -> " + details)

  def process_results(self, paths):
    dist = [row[0] for row in paths[1]] 
    time = [row[1] for row in paths[1]] 
    price = [row[2] for row in paths[1]] 

    self.get_min_max("distance",dist)
    self.get_min_max("time", time)
    self.get_min_max("price", price)
    self.get_shortest_and_longest(paths)
  
  def get_min_max(self, type, val):
    min_val = min(val)
    min_val_ind = val.index(min_val)
    print("\nMinimum " + type + ": " + str(min_val) + " " + str(paths[0][min_val_ind]))
    max_val = max(val)
    max_val_ind = val.index(max_val)
    print("Maximum " + type + ": " + str(max_val) + " " + str(paths[0][max_val_ind]))

  def filter_result(self, paths, by, min = 0, max = 0):
    dist = [row[0] for row in paths[1]] 
    time = [row[1] for row in paths[1]] 
    price = [row[2] for row in paths[1]] 

    if(by == 'distance'):
      a = numpy.array(dist)
    elif(by == 'time'):
      a = numpy.array(time)
    else:
      a = numpy.array(price)

    if(min >= 0 and max > 0):
      list = numpy.where(numpy.logical_and(a>=min, a<=max))
    elif(max == 0):
      list = numpy.where(a>=min)
    
    print("\n------------------------ Filtered ------------------------")
    summary = "By: " + by
    if(min != 0):
      summary += " | Min: " + str(min)
    if(max != 0):
      summary += " | Max: " + str(max)
    print(summary)

    for item in list[0]:
      filtered = str(paths[0][item])
      filtered += "[ Distance: " + str(paths[1][item][0]) + " | "
      filtered += "Time: " + str(paths[1][item][1]) + " | "
      filtered += "Price: " + str(paths[1][item][2]) + " ]"
      print(filtered)

  def print_paths(self, paths):
    for x, item in enumerate(paths[0]):
      details = "[ Distance: " + str(paths[1][x][0]) + " | "
      details += "Time: " + str(paths[1][x][1]) + " | "
      details += "Price: " + str(paths[1][x][2]) + " ]"
      print(str(item) + " -> " + details)


if __name__ == "__main__":
  routes = [
    ("Peirsophie", "The Hyster Depths", 3500, 44, 10500),
    ("Peirsophie", "Aelantir", 3750, 47, 7500),
    ("Peirsophie", "The Mirey", 5500, 69, 11000),
    ("Peirsophie", "The Mardeitsean Empire", 4000, 50, 8000),
    ("The Mardeitsean Empire", "Peirsophie", 4000, 50, 8000),
    ("The Mardeitsean Empire", "The Mirey", 1250, 16, 2500),
    ("The Mardeitsean Empire", "New Reksia", 5000, 63, 10000),
    ("The Mirey", "New Reksia", 2500, 32, 5000),
    ("The Mirey", "The Mardeitsean Empire", 1250, 16, 2500),
    ("The Mirey", "Peirsophie", 5500, 69, 11000),
    ("The Mirey", "Aelantir", 3000, 38, 6000),
    ("The Mirey", "Stiellbury", 2250, 29, 4500),
    ("New Reksia", "Stiellbury", 4750, 60, 9500),
    ("New Reksia", "The Mardeitsean Empire", 5000, 63, 10000),
    ("New Reksia", "The Mirey", 2500, 32, 5000),
    ("Stiellbury", "New Reksia", 4750, 60, 9500),
    ("Stiellbury", "The Unknown Graves", 2500, 32, 7500),
    ("Stiellbury", "Aelantir", 1500, 19, 3000),
    ("Stiellbury", "The Mirey", 2250, 29, 4500),
    ("Aelantir", "Stiellbury", 1500, 19, 3000),
    ("Aelantir", "The Hyster Depths", 2750, 35, 8250),
    ("Aelantir", "Peirsophie", 3750, 47, 7500),
    ("Aelantir", "The Mirey", 3000, 38, 6000),
    ("The Hyster Depths", "Aelantir", 2750, 35, 8250),
    ("The Hyster Depths", "Peirsophie", 3500, 44, 10500),
    ("The Unknown Graves", "Stiellbury", 2500, 32, 7500)
  ]
  start = "The Hyster Depths"
  end = "Stiellbury"

  # routes = [
  #   ("The Dragonmere", "Thunder Peaks", 4000, 14, 3200),
  #   ("Thunder Peaks", "The Dragonmere", 4000, 14, 3200),
  #   ("The Dragonmere", "Sembia", 6000, 20, 4800),
  #   ("Sembia", "The Dragonmere", 6000, 20, 4800),
  #   ("Thunder Peaks", "Sembia", 4000, 14, 3200),
  #   ("Sembia", "Thunder Peaks", 4000, 14, 3200),
  #   ("Thunder Peaks", "The Dalelands", 3000, 10, 2400),
  #   ("The Dalelands", "Thunder Peaks", 3000, 10, 2400),
  #   ("Sembia", "The Dalelands", 2000, 7, 1600),
  #   ("The Dalelands", "Sembia", 2000, 7, 1600),
  #   ("Sembia", "The Dragon Reach", 3500, 12, 2800),
  #   ("The Dragon Reach", "Sembia", 3500, 12, 2800),
  #   ("Sembia", "Pirate Isles", 5000, 17, 4000),
  #   ("Pirate Isles", "Sembia", 5000, 17, 4000),
  #   ("Pirate Isles", "The Dragon Reach", 4000, 14, 3200),
  #   ("The Dragon Reach", "Pirate Isles", 4000, 14, 3200),
  #   ("Pirate Isles", "Earthgas Mountains", 3000, 10, 2400),
  #   ("Earthgas Mountains", "Pirate Isles", 3000, 10, 2400),
  #   ("Pirate Isles", "Altumbel", 1000, 4, 800),
  #   ("Altumbel", "Pirate Isles", 1000, 4, 800),
  #   ("Altumbel", "Sea of Fallen Stars", 2000, 7, 1600),
  #   ("Sea of Fallen Stars", "Altumbel", 2000, 7, 1600),
  #   ("Sea of Fallen Stars", "Impiltur", 4000, 14, 3200),
  #   ("Impiltur", "Sea of Fallen Stars", 4000, 14, 3200),
  #   ("Sea of Fallen Stars", "Earthgas Mountains", 3000, 10, 2400),
  #   ("Earthgas Mountains", "Sea of Fallen Stars", 3000, 10, 2400),
  #   ("Earthgas Mountains", "Earthspur Mountains", 3000, 10, 2400),
  #   ("Earthspur Mountains", "Earthgas Mountains", 3000, 10, 2400),
  #   ("Earthspur Mountains", "Galena Mountains", 4000, 14, 3200),
  #   ("Galena Mountains", "Earthspur Mountains", 4000, 14, 3200),
  #   ("Earthspur Mountains", "The Dragon Reach", 15000, 50, 12000),
  #   ("The Dragon Reach", "Earthspur Mountains", 15000, 50, 12000),
  #   ("Galena Mountains", "Thar", 18000, 60, 14400),
  #   ("Thar", "Galena Mountains", 18000, 60, 14400),
  #   ("Thar", "The Moonsea", 1000, 4, 800),
  #   ("The Moonsea", "Thar", 1000, 4, 800),
  #   ("The Moonsea", "Dragonspine", 3000, 10, 2400),
  #   ("Dragonspine", "The Moonsea", 3000, 10, 2400),
  #   ("The Moonsea", "The Dragon Reach", 35000, 117, 28000),
  #   ("The Dragon Reach", "The Moonsea", 35000, 117, 28000),
  #   ("The Dragon Reach", "Cormanthor", 35000, 117, 28000),
  #   ("Cormanthor", "The Dragon Reach", 35000, 117, 28000),
  #   ("Cormanthor", "Dessertmouth", 45000, 150, 36000),
  #   ("Dessertmouth", "Cormanthor", 45000, 150, 36000),
  #   ("Dessertmouth", "The Dalelands", 25000, 84, 20000),
  #   ("The Dalelands", "Dessertmouth", 25000, 84, 20000)
  # ]
  # start = "Thar"
  # end = "Pirate Isles"
  
  
  # routes = [
  #   ("The Scrapyard", "The Gate", 700, 9, 35),
  #   ("The Gate", "The Scrapyard", 700, 9, 35),
  #   ("The Scrapyard", "The Walk", 100, 2, 5),
  #   ("The Walk", "The Scrapyard", 100, 2, 5),
  #   ("The Scrapyard", "Scepter Streets", 4300, 54, 215),
  #   ("Scepter Streets", "The Scrapyard", 4300, 54, 215),
  #   ("The Gate", "The Walk", 7700, 97, 385),
  #   ("The Walk", "The Gate", 7700, 97, 385),
  #   ("The Gate", "Windsdale Garden", 7400, 93, 370),
  #   ("Windsdale Garden", "The Gate", 7400, 93, 370),
  #   ("The Gate", "Dondorian Plaza", 6200, 78, 310),
  #   ("Dondorian Plaza", "The Gate", 6200, 78, 310),
  #   ("The Gate", "Bricksmoor District", 4700, 59, 235),
  #   ("Bricksmoor District", "The Gate", 4700, 59, 235),
  #   ("Bricksmoor District", "Dondorian Plaza", 3800, 48, 190),
  #   ("Dondorian Plaza", "Bricksmoor District", 3800, 48, 190),
  #   ("Bricksmoor District", "Guilds Lane", 400, 5, 20),
  #   ("Guilds Lane", "Bricksmoor District", 400, 5, 20),
  #   ("Guilds Lane", "Dondorian Plaza", 3300, 42, 165),
  #   ("Dondorian Plaza", "Guilds Lane", 3300, 42, 165),
  #   ("Guilds Lane", "Eastern Docks", 2800, 35, 140),
  #   ("Eastern Docks", "Guilds Lane", 2800, 35, 140),
  #   ("Eastern Docks", "Western Docks", 6700, 84, 335),
  #   ("Western Docks", "Eastern Docks", 6700, 84, 335),
  #   ("Eastern Docks", "Windsdale Garden", 6500, 82, 325),
  #   ("Windsdale Garden", "Eastern Docks", 6500, 82, 325),
  #   ("Western Docks", "Old Town", 3600, 45, 180),
  #   ("Old Town", "Western Docks", 3600, 45, 180),
  #   ("Old Town", "Windsdale Garden", 6100, 77, 305),
  #   ("Windsdale Garden", "Old Town", 6100, 77, 305),
  #   ("Old Town", "Cathedral District", 3700, 47, 185),
  #   ("Cathedral District", "Old Town", 3700, 47, 185),
  #   ("Old Town", "Scepter Streets", 4500, 57, 225),
  #   ("Scepter Streets", "Old Town", 4500, 57, 225),
  #   ("Cathedral District", "The Walk", 1800, 23, 90),
  #   ("The Walk", "Cathedral District", 1800, 23, 90),
  #   ("Cathedral District", "Windsdale Garden", 2300, 29, 115),
  #   ("Windsdale Garden", "Cathedral District", 2300, 29, 115),
  #   ("Windsdale Garden", "Dondorian Plaza", 1300, 17, 65),
  #   ("Dondorian Plaza", "Windsdale Garden", 1300, 17, 65)
  # ]
  # start = "The Gate"
  # end = "Eastern Docks"
  


  route_graph = Graph(routes)
  
  print(f"\nCheck all paths between: {start} and {end}")
  paths = route_graph.get_paths(start,end)
  print("Total Paths: ", len(paths[0]))
  print(f"All paths between: {start} and {end}:")
  route_graph.print_paths(paths)
  print("\n************************ Summary ************************")
  route_graph.process_results(paths)

  #Ems'
  route_graph.filter_result(paths, "distance", 0, 15000)
  route_graph.filter_result(paths, "time", 90, 240)
  route_graph.filter_result(paths, "price", 0, 40000)

  # Windsdale
  # route_graph.filter_result(paths, "distance", 35000, 125000)
  # route_graph.filter_result(paths, "time", 0, 600)
  # route_graph.filter_result(paths, "price", 50000, 0)

  #Central
  # route_graph.filter_result(paths, "distance", 28000, 0)
  # route_graph.filter_result(paths, "time", 0, 200)
  # route_graph.filter_result(paths, "price", 800, 1000)
```
